# Dataset Understanding & Exploratory Data Analysis (EDA)
This notebook explores the `application_record.csv` and `credit_record.csv` datasets.

## Steps:
1. Load and understand the data.
2. Check for missing values and data types.
3. Perform Exploratory Data Analysis (EDA) with visualizations.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
plt.style.use("ggplot")
```

```python
# Load datasets
application_df = pd.read_csv('application_record.csv')
credit_df = pd.read_csv('credit_record.csv')

# Display first few rows of each dataset
application_df.head(), credit_df.head()
```

```python
# Dataset information
application_df.info()
credit_df.info()
```

```python
# Check missing values
application_df.isnull().sum(), credit_df.isnull().sum()
```

## Exploratory Data Analysis (EDA)
Now, let's visualize key aspects of the dataset.


```python
# Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x=application_df["CODE_GENDER"])
plt.title("Distribution of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Income Distribution
plt.figure(figsize=(8, 4))
sns.histplot(application_df["AMT_INCOME_TOTAL"], bins=50, kde=True)
plt.title("Distribution of Total Income")
plt.xlabel("Total Income")
plt.ylabel("Count")
plt.xlim(0, application_df["AMT_INCOME_TOTAL"].quantile(0.99))  # Removing extreme outliers for better visualization
plt.show()
```

```python
# Credit Status Distribution
plt.figure(figsize=(8, 4))
sns.countplot(x=credit_df["STATUS"], order=credit_df["STATUS"].value_counts().index)
plt.title("Distribution of Credit Status")
plt.xlabel("Credit Status")
plt.ylabel("Count")
plt.show()
```

```python
# Relationship between Income and Education
plt.figure(figsize=(12, 6))
sns.boxplot(x="NAME_EDUCATION_TYPE", y="AMT_INCOME_TOTAL", data=application_df)
plt.xticks(rotation=45)
plt.title("Income Distribution by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Total Income")
plt.ylim(0, application_df["AMT_INCOME_TOTAL"].quantile(0.99))  # Remove extreme outliers for better visualization
plt.show()
```

## Summary
- **Gender**: Distribution between male and female applicants.
- **Income**: Most applicants earn below 1,000,000 per year.
- **Credit Status**: Majority of applicants have no overdue payments.
- **Education & Income**: Higher education levels tend to correlate with higher income.

This concludes the **exploratory data analysis** for the dataset.
